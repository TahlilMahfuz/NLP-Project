# Dataset Preprocessing

## MEQSum
Maps into input and target

```python
import re
from datasets import DatasetDict

class MeQSumPreprocessor:
    """
    Preprocessor for MeQSum dataset using Template Method pattern.
    Transforms (CHQ, Summary) into a task-description dataset:
    'verbose question → short question'.
    """

    def __init__(self, dataset: DatasetDict):
        self.dataset = dataset

    def preprocess(self) -> DatasetDict:
        return self.dataset.map(self._process_batch, batched=True)

    def _process_batch(self, batch):
        inputs = [self._format_input(chq) for chq in batch["CHQ"]]
        targets = [self._format_target(summary) for summary in batch["Summary"]]
        return {"input": inputs, "target": targets}

    def _format_input(self, text: str) -> str:
        """Format verbose CHQ into standardized input prompt."""
        text = self._clean_text(text)
        return f"Task: Convert verbose patient questions into short medical questions.\n\nInput: {text}\nOutput:"

    def _format_target(self, text: str) -> str:
        """Clean target summary."""
        return self._clean_text(text)

    def _clean_text(self, text: str) -> str:
        """Remove SUBJECT:, MESSAGE:, newlines, and excessive whitespace."""
        if not isinstance(text, str):
            return ""
        text = re.sub(r"(SUBJECT:|MESSAGE:)", "", text, flags=re.IGNORECASE)
        text = re.sub(r"\s+", " ", text)
        return text.strip()
```

```python
processed_meqsum = MeQSumPreprocessor(meqsum).preprocess()
```

# MIMIC-CXR
maps into input and target

```python
import re
from datasets import DatasetDict

class MIMICCXRPreprocessor:
    """
    Preprocessor for MIMIC-CXR dataset using Template Method pattern.
    Transforms (findings → impression) into a standardized task description.
    """

    def __init__(self, dataset: DatasetDict):
        self.dataset = dataset

    def preprocess(self) -> DatasetDict:
        return self.dataset.map(self._process_batch, batched=True)

    def _process_batch(self, batch):
        inputs = [self._format_input(f) for f in batch["findings"]]
        targets = [self._format_target(i) for i in batch["impression"]]
        return {"input": inputs, "target": targets}

    def _format_input(self, text: str) -> str:
        """Format findings text into instruction-style input."""
        text = self._clean_text(text)
        return (
            "Task: Summarize verbose chest X-ray findings into a short clinical impression.\n\n"
            f"Findings: {text}\nImpression:"
        )

    def _format_target(self, text: str) -> str:
        """Format impression as target summary."""
        return self._clean_text(text)

    def _clean_text(self, text: str) -> str:
        """Normalize whitespace and strip special markers."""
        if not isinstance(text, str):
            return ""
        text = re.sub(r"\s+", " ", text)
        return text.strip()
```

```python
processed_cxr = MIMICCXRPreprocessor(cxr).preprocess()
```

# Open-I Data Preprocessing

```python
import re
from datasets import DatasetDict

class OpenIPreprocessor:
    """
    Preprocessor for Open-i chest X-ray dataset.
    Converts (findings → impression) into a standardized instruction-style dataset.
    """

    def __init__(self, dataset: DatasetDict, use_indication: bool = True, use_comparison: bool = True):
        self.dataset = dataset
        self.use_indication = use_indication
        self.use_comparison = use_comparison

    # --- Template method ---
    def preprocess(self) -> DatasetDict:
        return self.dataset.map(self._process_batch, batched=True)

    # --- Hooks ---
    def _process_batch(self, batch):
        inputs = []
        targets = []

        for f, i, ind, comp in zip(
            batch["findings"], batch["impression"], batch["indication"], batch["comparison"]
        ):
            inputs.append(self._format_input(f, ind, comp))
            targets.append(self._format_target(i))

        return {"input": inputs, "target": targets}

    def _format_input(self, findings: str, indication: str, comparison: str) -> str:
        """Build instruction text with optional context."""
        findings = self._clean_text(findings)

        context_parts = []
        if self.use_indication and indication:
            context_parts.append(f"Indication: {self._clean_text(indication)}")
        if self.use_comparison and comparison:
            context_parts.append(f"Comparison: {self._clean_text(comparison)}")

        context = "\n".join(context_parts)

        return (
            "Task: Summarize verbose chest X-ray findings into a concise clinical impression.\n\n"
            + (context + "\n" if context else "")
            + f"Findings: {findings}\nImpression:"
        )

    def _format_target(self, impression: str) -> str:
        return self._clean_text(impression)

    def _clean_text(self, text: str) -> str:
        """Normalize whitespace and strip artifacts."""
        if not isinstance(text, str):
            return ""
        text = re.sub(r"\s+", " ", text)
        text = re.sub(r"XXXX", "", text)  # anonymized placeholders
        return text.strip()
```

```python
processed_openi = OpenIPreprocessor(openi).preprocess()
```
